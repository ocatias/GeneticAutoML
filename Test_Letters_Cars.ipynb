```python
import pandas as pd

import importlib
import timeit

import GenAlgo_1 as GA

from sklearn.model_selection import train_test_split

importlib.reload(GA)

from sklearn.preprocessing import LabelEncoder

def labelEncoding(data):
    data_labelenc = data.copy()

    #Find columns with categorical values
    s = (data_labelenc.dtypes == 'object')
    object_cols = list(s[s].index)

    #Replace those by labels
    for col in object_cols:
        labels_df = pd.DataFrame(data_labelenc, columns=[col])
        labelencoder = LabelEncoder()
        data_labelenc[col] = labelencoder.fit_transform(labels_df[col])

    return data_labelenc
```

```python
duration_in_minutes = 60
individuals_GA_letter = 50
individuals_GA_car = 100
```

# LETTER DATASET

```python
filepath = "letter-recognition.data"
data = pd.read_csv(filepath, header =None) 
data = data.head(5000)

y = data[0]
X = data.drop(axis=1, labels=0)


from sklearn.preprocessing import LabelEncoder 

lenc = LabelEncoder()
y = lenc.fit_transform(y)

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 0)
```

```python
importlib.reload(GA)

start = timeit.default_timer()

GenAutoML = GA.GenAutoML(n_jobs=-1,nr_individuals=individuals_GA_letter,nr_iterations=1000000, timelimit=60*duration_in_minutes, 
                         equal_distribution = False,
                         verbose = 11, crossvalidate = True, cv = 5)
GenAutoML.fit(X_train,y_train)
stop = timeit.default_timer()

print('\nRuntime: ', stop - start, 's')  
```

```python
predictions = GenAutoML.predict(X_test,y_test)
```

```python
from tpot import TPOTClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

labelencoder = LabelEncoder()

pipeline_optimizer = TPOTClassifier(max_time_mins = duration_in_minutes, generations = None, cv=5,
                                    random_state=42, verbosity=2, scoring = 'f1_macro')

pipeline_optimizer.fit(X_train, labelencoder.fit_transform(y_train))
print(pipeline_optimizer.score(X_test, labelencoder.transform(y_test)))
```

```python
import h2o
from h2o.automl import H2OAutoML
h2o.init()

# https://docs.h2o.ai/h2o/latest-stable/h2o-docs/automl.html#code-examples

h2o_aml = H2OAutoML(seed=42, max_runtime_secs=duration_in_minutes*60, max_models=20, max_runtime_secs_per_model=180, nfolds=5)

X_train_copy = X_train.copy()
X_test_copy = X_test.copy()

X_train_copy.insert(value=y_train, loc=0, column=0)
X_test_copy.insert(value=y_test, loc=0, column=0)

train = h2o.H2OFrame(X_train_copy)
valid = h2o.H2OFrame(X_test_copy)


x = train.columns
y = "0"
x.remove(y)

train[y] = train[y].asfactor()
valid[y] = valid[y].asfactor()

start = timeit.default_timer()

h2o_aml.train(x = x,
                   y = y,
                   training_frame = train,
                   validation_frame = valid)

stop = timeit.default_timer()

print('\nRuntime: ', stop - start, 's')  
```

```python
from sklearn.metrics import f1_score

leaderbord = h2o_aml.leaderboard
#print(leaderbord)
leader = h2o_aml.leader
#print(leader)
perf = leader.model_performance(valid)
print(perf)
pred = leader.predict(valid)
#print(valid[0])
#print(pred[0])

print("Performance of H2O Auto-ML Classifier")
print(f1_score(h2o.as_list(valid[0]), h2o.as_list(pred[0]), average='macro'))
```
